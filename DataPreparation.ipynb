## Cleaning Hashtags column having empty lists and nulls

```python
df['Hashtags'].head()
```

```python
df['Hashtags'] = df['Hashtags'].apply(lambda x: 0 if x=='[]' else x)
```

```python
df['Hashtags'].head()
```

```python
df['Hashtags']=df['Hashtags'].fillna(0)
```

```python
df['Hashtags'].head()
```

```python
df['Hashtags_count']=df['Hashtags'].str.split().str.len()
```

```python
df['Hashtags_count'].value_counts()
```

```python
df['Hashtags_count']=df['Hashtags_count'].fillna(0)
```

```python
df.isna().sum()
```

## Cleaning is_media

```python
df['is_media'].head()
```

```python
df['is_media']=df['is_media'].fillna(0)
```

```python
df['is_media']=df['is_media'].replace(0,'None')
```

```python
df['is_media'].value_counts()
```

## Month, Date, Year, Time and Recency from Datetime stamp

```python
#Get Month as names from datetime stamp
df['Month']=pd.DatetimeIndex(df['created_at']).month
dayOfMonth={1: 'January', 2: 'February', 3: 'March', 4:'April', 5:'May', 6:'June', 7:'July', 8:'August', 9:'September', 
           10:'October', 11:'November', 12:'December'}
df['Month']=df['Month'].map(dayOfMonth)
```

```python
#Date of Month
df['day_month']=pd.DatetimeIndex(df['created_at']).day
```

```python
#Year
df['Year']=pd.DatetimeIndex(df['created_at']).year
```

```python
#Time
df['time']=pd.DatetimeIndex(df['created_at']).time
```

```python
#Get day of week as mon tue etc from datetime
df['day_week']=pd.DatetimeIndex(df['created_at']).dayofweek
dayOfWeek={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
df['day_week'] = df['day_week'].map(dayOfWeek)
```

```python
del df['Team_Member']
```

```python
df['Tweet_URL.1']=df['Tweet_URL.1'].fillna('Unknown')
```
